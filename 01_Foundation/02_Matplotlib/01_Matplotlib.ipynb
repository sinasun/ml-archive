# Matplotlib

Just a quick overview of most used matplotlib functions.

```python
import numpy as np
import matplotlib.pyplot as plt
```

## 1 Plotting Points

### 1.1 Basic
Using ``plt.scatter()`` we can plot the points:
``plt.scatter(x, y)``

    

```python
np.random.seed(1) # Setting seed so we get same output
x_data = np.random.random((10,)) # Random array of 10, range of input are between 0 and 1
y_data = np.random.random((10,)) # Random array of 10, range of input are between 0 and 1

for i in range(x_data.shape[0]):
    print(f"x: {x_data[i]}, y: {y_data[i]}")

plt.scatter(x_data, y_data)
```

### 1.2 Color

By passing ``color`` argument to scatter we can define the color of the points. If we pass a list of colors, the color is going to be for each point, or we can pass a color for each set of points. See below for examples:

```python
np.random.seed(2) # Setting seed so we get same output
x_data = np.random.random((10,)) # Random array of 10, range of input are between 0 and 1
y_data = np.random.random((10,)) # Random array of 10, range of input are between 0 and 1


plt.scatter(x_data, y_data, color='red')
```

```python
np.random.seed(3) # Setting seed so we get same output
x_data = np.random.random((10,)) # Random array of 10, range of input are between 0 and 1
y_data = np.random.random((10,)) # Random array of 10, range of input are between 0 and 1
colors = np.random.choice(['pink', 'red', 'blue', 'green', 'black'], 10) # Create random array of 10 from the given list

plt.scatter(x_data, y_data, color=colors)
```

### 1.3 Size

By passing `s` to the ``scatter`` you can define the size of the elements. It works same as `color`, (the way it applies to points). Example:

```python
np.random.seed(4)
x_data = np.random.random((10,))
y_data = np.random.random((10,))
sizes = np.random.uniform(low=1, high=500, size=10) # Random 10 numbers from 1 to 5

plt.scatter(x_data, y_data, s=sizes)
```

## 2 Plotting Function

Using ``plt.plot(x, y)`` you can make points that are connected together. 

``color`` defines the color of the function.

Note: With ``plt.plot(x, y, marker='o')`` you can plot points too, but it's not recommeneded as it returns a line object and has less modification in compare to ``plt.scatter(x, y)``

```python
np.random.seed(1)
x_data = np.random.random((10,))
x_data = np.sort(x_data) # Sorting our x data so we get a better function

y_data = np.random.random((10,))


plt.plot(x_data, y_data, color='green')
```

## 3 Legends and Label

You can assign a label to ``plt.scatter()`` and ``plt.plot()``. Using legends, you can show that each element is for what label. Also, you can assign labels for your X and Y axis.

```python
# Random Points

x_points = np.random.uniform(0, 10, 10) # 10 random numbers in range of 0 to 10
y_points = np.random.uniform(-1, 1, 10) # 10 random numbers in ragne of -1 to 1
plt.scatter(x_points, y_points, color='green', label='Random Points') # Plot the points

# y = sin(x) function
def f_x(x):
    return np.sin(x)

x_data = np.random.uniform(0, 10, 1_000) # Random 1,000 numbers from 1 to 10
x_data = np.sort(x_data) # Sorting our x data so we get a better function

y_data = f_x(x_data) # Find y = sin(x) for each of them

plt.plot(x_data, y_data, color='red', label="y = sin(x)")


plt.xlabel('X')
plt.ylabel('Y')
plt.legend()

plt.show()
```
